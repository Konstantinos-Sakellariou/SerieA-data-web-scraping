```
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```
teams = ["atalanta", "benevento", "bologna", "cagliari", "crotone", "fiorentina", "genoa", "hellas-verona", "inter", "juventus",
         "lazio", "milan", "napoli", "parma", "roma", "sampdoria", "sassuolo", "spezia", "torino", "udinese"]

number_of_players_per_team = [30, 29, 36, 32, 33, 32, 32, 39, 36, 36, 37, 38, 28, 42, 43, 27, 32, 32, 29, 30]
```

```
k = 0
team_set = {'team'}
serieA_data = pd.DataFrame( columns = ["Number", "Player Name", "Position", "Appearences", "Minutes", "Yellow Cards",
                                         "YellowToRedCards", "Red Cards", "KM Run", "Goal_chances", "Assists", "Shots", "Team"])


for j in teams:

  page = 'http://www.legaseriea.it/en/serie-a/teams/team/players-stats'
  page = ' '.join( j if i in team_set else i for i in page.split("/"))
  page = page.replace(' ',"/")
  
  url = requests.get(page)
  soup = BeautifulSoup(url.content,"html.parser")

  table = soup.find("table",{"class":"tabella tabella_swipe"})
  tag = table.findAll('td')

  player_number = []
  player_name = []
  player_position = []
  player_appearences = []
  player_minutes = []
  player_goals = []
  player_yellow_cards= []
  player_yellow_to_red_cards = []
  player_red_cards = []
  player_km_run = []
  player_goal_chances = []
  player_assists = []
  player_shots = []
        
  m = 0
  x = number_of_players_per_team[k]

  for i in range(x):
    
    player_number.append(tag[0+m].text.replace("\n","").strip())
    player_name.append(tag[1+m].text.replace("\n","").strip())
    player_position.append(tag[2+m].text.replace("\n","").strip())
    player_appearences.append(tag[3+m].text.replace("\n","").strip())
    player_minutes.append(tag[4+m].text.replace("\n","").strip())
    player_goals.append(tag[5+m].text.replace("\n","").strip())
    player_yellow_cards.append(tag[6+m].text.replace("\n","").strip())
    player_yellow_to_red_cards.append(tag[7+m].text.replace("\n","").strip())
    player_red_cards.append(tag[8+m].text.replace("\n","").strip())
    player_km_run.append(tag[9+m].text.replace("\n","").strip())
    player_goal_chances.append(tag[10+m].text.replace("\n","").strip())
    player_assists.append(tag[11+m].text.replace("\n","").strip())
    player_shots.append(tag[12+m].text.replace("\n","").strip())
    m = m + 13

  team_data = pd.DataFrame({"Number": player_number, "Player Name": player_name, "Position": player_position, "Appearences": player_appearences, "Minutes": player_minutes, "Yellow Cards": player_yellow_cards,
                          "YellowToRedCards": player_yellow_to_red_cards, "Red Cards": player_red_cards, "KM Run": player_km_run, "Goal_chances": player_goal_chances, "Assists": player_assists, "Shots": player_shots,
                          "Team": j})
  
  serieA_data = pd.concat([serieA_data,team_data])
  k += 1
  
```
